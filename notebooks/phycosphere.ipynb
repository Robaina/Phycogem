# Candidate objective functions

Candidate objective functions to explore, which may actually require full community modelling, rather than only the phycosphere:

1. Metabolic Dominance Detection: This function aims to quantify the metabolic dominance of the phycosphere during the early stages of a HAB event. It could be defined as the ratio of the total metabolic flux of the dominant HAB species to the total metabolic flux of the entire microbial community. A significant increase in this ratio could indicate the onset of a HAB event.

2. Early Warning Signal Detection: This function aims to identify drastic shifts in metabolic fluxes that could serve as early warning signals for a HAB event. It could be defined as the rate of change of the total metabolic flux of the microbial community, with a significant increase indicating a potential HAB event.

3. Ecological Health Metric Prototyping: This function aims to develop ecological health metrics that can effectively monitor coastal ecosystems. It could be defined as a composite score of the above two functions, with a significant deviation from the baseline indicating a potential threat to the ecological health of the ecosystem.

But we need to iterate upon this several times, and also brainstorm ideas with the rest of the sci-team!

It looks like a large part of the interactions between bacteria and microalgae at the onset of a HAB event are related to cofactors/vitamins and iron or other trace elements in exchange of DOC [1, 2]. Thus, I would expect a shift in metabolic fluxes favoring the production of secreted DOCs by the microalgae [3, 4] and a marked increase in the production and secretion of cofactors/vitamins and siderophores by the bacterial community, as well as the pathways involved in uptaking and processing the diverse DOCs. Other studies have pointed to Quorum Sensing-related pathways during HAB events [5]. These changes could serve as a metabolic, system-wise, signal to detect the onset of a HAB event.

We could integrate metatranscriptomic datasets of HAB events [6, 7] into the community/phycosphere metabolic model to find the closest metabolic network state matching the metatranscriptome (this probably will be a collection of optimal states, think of flux sampling constrained by transcript data). By comparing these states with baseline states (e.g, metaTs outside of a HAB event) we could learn community-wide metabolic signatures that are characteristic of the onset of a HAB event.
